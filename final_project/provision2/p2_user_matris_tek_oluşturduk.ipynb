```python
# Null olan değerlerin hangi grupta olduğunu bulup, o grubun ortalamasını o değere atama
for index, row in veri[veri['ShippingCost'].isnull()].iterrows():
    description = row['Description']  # Null değerin bulunduğu satırın Descriptionu
    mean_value = veri[veri['Description'] == description]['ShippingCost'].mean()  # O description için ortalama hesaplama
    veri.at[index, 'ShippingCost'] = mean_value  # Null değeri ortalama ile doldur
```

```python
veri['ReturnStatus'] = veri['ReturnStatus'].map({'Returned': 0, 'Not Returned': 1})
```

```python
veri.head()
```

```python
# Normalize işlemleri

# Her bir Description için normalize edilmiş UnitPrice_Contribution hesaplama
def calculate_unitprice_contribution(group):
    group['UnitPrice_Contribution'] = group['UnitPrice'] / group['UnitPrice'].max()
    return group
# Gruplayarak UnitPrice_Contribution hesaplama
veri = veri.groupby('Description', group_keys=False).apply(calculate_unitprice_contribution)

veri['Quantity_Contribution'] = veri['Quantity'] / 50
veri['Discount_Contribution'] = 1 - (veri['Discount'] / 2)+0.1
veri['Shipping_Contribution'] = 1 - (veri['ShippingCost'] - 5) / (30 - 5)+0.1
veri['Return_Contribution'] = veri['ReturnStatus']

# OrderPriority puanlama
priority_mapping = {'High': 1.0, 'Medium': 0.5, 'Low': 0.2}
veri['Priority_Contribution'] = veri['OrderPriority'].map(priority_mapping)

# Rating hesaplama
weights = [-0.2, 0.05, -0.2, 0.2, 0.2, 0.2]
veri['Rating'] = (
    weights[0] * veri['UnitPrice_Contribution'] +
    weights[1] * veri['Quantity_Contribution'] +
    weights[2] * veri['Discount_Contribution']+0.2+
    weights[3] * veri['Shipping_Contribution'] +
    weights[4] * veri['Return_Contribution'] +
    weights[5] * veri['Priority_Contribution']
)
veri.head(10)
```

```python
veri.describe()
```

```python
veri['Rating'] = veri['Rating'] + 0.2
```

```python
veri.describe()
```

```python
df = pd.DataFrame(veri)

# User-Product matrisini oluşturma (ortalama ile toplulaştırma)
user_product_matrix = df.pivot_table(index='CustomerID', columns='Description', values='Rating', aggfunc='mean')
```
